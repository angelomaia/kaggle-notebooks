# Importing libraries and loading data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

!pip install mplfinance

import mplfinance as mpf

data = pd.read_csv('/kaggle/input/roblox-stock-pricing-2021-2023/RBLX.csv')
```

```python
data.head()
```

## Let's set the 'Date' column as the index in order to use mplfinance plot function

```python
data.set_index('Date', inplace=True)
data.index = pd.to_datetime(data.index)
```

### Now let's plot the OHLC (Open, High, Low, Close) candlestick graph

```python
mpf.plot(data, type='candle', figsize=(18,6))
```

### It's hard to look at the candles in daily timeframe. Let's transform the OHLC values to a weekly timeframe

```python
weekly_data = data.resample('W', closed='right', label='right').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
})
```

```python
weekly_data.head()
```

```python
weekly_data.index = pd.to_datetime(weekly_data.index)

mpf.plot(weekly_data, type='candle', figsize=(18,6))
```

### Now let's add Moving Averages to the graph

```python
mpf.plot(weekly_data, type='candle', mav=(9,18,36), figsize=(18,6))
```

### We can also add Volume to the graph

```python
mpf.plot(weekly_data, type='candle', mav=(9,18,36), figsize=(18,6), volume=True)
```

### This graph tells us that there has been a sudden price rise to ~140 USD in late 2021. Such rise was quickly followed by a gradual decline to a minimum of ~30 USD, leading the stock to a less volatile behaviour (even under high volume trading weeks). This pattern lasts until nowadays.
