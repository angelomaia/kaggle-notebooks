```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/zomato-restaurants-dataset-for-metropolitan-areas/zomato_dataset.csv')
```

## Let's open up this dataset and visualize the data.

```python
print(df.shape)
print(df.columns)
print(df.describe())
```

```python
sns.histplot(df['Dining Rating'], bins=12, kde=True)
plt.title('Distribution of Dining Ratings')
plt.show()

sns.histplot(df['Delivery Rating'], bins=16, kde=True)
plt.title('Distribution of Delivery Ratings')
plt.show()
```

### People are apparently more likely to give a better rating for deliveries rather than restaurants.

```python
sns.scatterplot(data=df, x='Dining Rating', y='Dining Votes')
plt.title('Dining Rating vs. Dining Votes')
plt.show()

sns.scatterplot(data=df, x='Delivery Rating', y='Delivery Votes')
plt.title('Delivery Rating vs. Delivery Votes')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.countplot(data=df, y='Cuisine ')
plt.title('Count of Restaurants by Cuisine')
plt.show()
```

### Beverages are crushin' it.

```python
counts = df['Best Seller'].value_counts()
proportions = counts / counts.sum()

threshold = 0.05
small_proportions = proportions[proportions < threshold]
proportions['Others'] = small_proportions.sum()
proportions = proportions[proportions >= threshold]

colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99FF', '#FFFF99',
          '#99FFFF', '#FFCCCC', '#CCCCFF', '#99CC99', '#CC99CC', '#FFCCFF']

plt.figure(figsize=(8, 8))
plt.pie(proportions, labels=proportions.index, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'white'})

plt.axis('equal')

plt.title('Proportion of Best Sellers')
plt.legend(loc='center left', bbox_to_anchor=(1.4, 0.5))

plt.show()
```

### The best bestsellers are... bestsellers!

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Delivery Rating'], df['Prices'])
plt.xlabel('Delivery Rating')
plt.ylabel('Prices')
plt.title('Scatter Plot: Delivery Rating vs Prices')
plt.show()
```
